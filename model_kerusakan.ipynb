```python
import pandas as pd
```

```python
kerusakan = pd.read_csv('data_kerusakan.csv')
```

```python
kerusakan
```

```python
data = kerusakan.copy()
```

```python
target = 'Kerusakan'
```

```python
data.head()
```

```python
target_mapper = {'Internet Limit':0, 'Router Reset':1, 'Network Cable Unplugged':2, 'Router/Switch Rusak':3}
```

```python
def target_encode(val):
  return target_mapper[val]
```

```python
data['Kerusakan'] = data['Kerusakan'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('Kerusakan', axis=1)
Y = data['Kerusakan']
```

```python
X
```

```python
Y
```

```python
from sklearn.svm import SVC
model = SVC(probability=True).fit(X, Y)
```

```python
import pickle
pickle.dump(model, open('model-kerusakan.pkl', 'wb'))
```
